## Informações Estatísticas da Violência no Rio de Janeiro

Iremos entrar mais a fundo na analise, a partir da nossa breve contextualizacao dos dados , para entender melhor a situacao da segurança do estado RJ.



```
#resumo estatístico 
df.describe()
```

```
df['roubo_veiculo'].mean() , df['furto_veiculos'].mean() ,df['recuperacao_veiculos'].mean()
```

```
#porcentagem de carros recuperados em relação aos carros roubados + carros furtados
(df['recuperacao_veiculos'].sum())/(df['roubo_veiculo'].sum() + df['furto_veiculos'].sum())*100 
```
